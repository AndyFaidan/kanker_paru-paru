## membuat model training

model=LogisticRegression() adalah sintaks yang digunakan untuk membuat objek LogisticRegression sebagai model yang akan digunakan untuk melakukan klasifikasi pada data. Objek ini akan dilatih pada data yang telah diproses sebelumnya dan akan menghasilkan prediksi pada data baru berdasarkan pembelajaran yang telah dilakukan. Pada sintaks ini, model adalah nama variabel yang digunakan untuk menyimpan objek LogisticRegression yang dibuat.

```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

## evaluasi model

Evaluasi adalah proses pencatatan hasil pembelajaran menggunakan metrik evaluasi yang berbeda untuk memahami kinerja model. Evaluasi model juga berperan dalam pemantauan dan pemilihan model terbaik dengan melihat score dari data test dan data train masing-masing algoritma. 

Kode dibawah menghasilkan prediksi model pada data latih (X_train) dan membandingkannya dengan target aktual dari data latih (Y_train) untuk menghitung akurasi model pada data latih (training_data_accuracy). Prediksi model dihasilkan dengan memanggil fungsi predict() pada objek model yang telah dibuat sebelumnya. Kemudian, hasil prediksi dibandingkan dengan target aktual menggunakan fungsi accuracy_score() dari library scikit-learn untuk menghitung akurasi model.

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```
